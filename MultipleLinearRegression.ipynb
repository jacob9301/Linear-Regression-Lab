## Multiple linear regression

using multiple linear regression on the Boston data set contained within the ISLP package.<br />
the goal of multiple linear regression is to create a function for which $Y \approx \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \ldots + \beta_j X_j$. where the values of $\beta_i$ minimise the residual sum of squares (RSS), $X_i$ are our predictors and $Y$ is our response.
